지난 1년간 아파트의 평단가는 상승했을까?

```python
def get_Average_unit_price(yymm, data):
  sumOfPrices = 0.0
  sumOfSizes = 0.0
  trade_count = 0
  average_unit_price = 0.0
  for i in range(0, len(data)):
    if(data[i][0] == yymm):
      trade_count = trade_count + 1
      sumOfPrices = sumOfPrices + data[i][2]
      sumOfSizes = sumOfSizes + data[i][1]

  average_unit_price = sumOfPrices / sumOfSizes * 3.3

  return average_unit_price, trade_count
```

```python
data = []
for i in range(0, len(df)):
  area = float(df.loc[i][5]) # 전용면적
  year_month = int(df.loc[i][6]) # 계약년월
  price = float(str(df.loc[i][8]).replace(",", "")) # 거래금액

  # 평단가 구하기 price / area * 3.3 <- 한평 기준
  unit_price = float(price / area * 3.3)
  # print(area, year_month, price, unit_price)

  if (len(str(df.loc[i][12])) > 5):
    pass
  else:
    data.append([year_month, area, price, unit_price])

print(len(data))
```

```python
data_monthly = []
for i in range(0, 13):
  yymm = 2022
  if i >= 10:
    yymm = 2023
  yymm = yymm * 100
  if i >= 10:
    i = i - 12
  yymm = yymm + (i + 3)
  
  average_unit_price, trade_count = get_Average_unit_price(yymm, data)
  data_monthly.append([average_unit_price, trade_count])

for i in data_monthly:
  print(i)
```
