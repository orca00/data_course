```python
import numpy as np
import pandas as pd

# 소수점 첫 번째 자리까지 표시
pd.set_option('precision', 1)

# 데이터프레임 생성
df_left = pd.DataFrame({'색': ['red', 'orange', 'yellow', 'green', 'blue'],
                        '가격': [1000, 2000, 3000, 4000, 5000]},
                       index=[9, 1, 2, 3, 4])

df_right = pd.DataFrame({'색': ['yellow', 'green', 'blue', 'navy', 'purple'],
                         '이름': ['노란색', '초록색', '파란색', '남색', '보라색']},
                        index=[3, 4, 5, 6, 7])

# 데이터프레임 df_left & df_right 확인
display([df_left, df_right], ['df_left', 'df_right'])
```

```python
# Inner Join
df_inner = pd.merge(df_left, df_right, on="색", how="inner")

# 데이터 프레임 df_left & df_right, df_inner 확인
display([df_left, df_right, df_inner], ['df_left', 'df_right', 'df_inner'])
```

```python
# Outer Join
df_outer = pd.merge(df_left, df_right, on="색", how="outer")

# 데이터프레임 df_left & df_right & df_outer 확인
display([df_left, df_right, df_outer], ['df_left', 'df_right', 'df_outer'])
```

```python
# Left Join
df_left_join = pd.merge(df_left, df_right, on="색", how="left")

# 데이터프레임 df_left & df_right, df_left_join 확인
display([df_left, df_right, df_left_join], ['df_left', 'df_right', 'df_left_join'])
```

```python
# Right Join
df_right_merge = pd.merge(df_left, df_right, on="색", how="right")

# 데이터프레임 df_left & df_right, df_right_merge 확인
display([df_left, df_right, df_right_merge], ['df_left', 'df_right', 'df_right_merge'])
```

```python
# Right_Join - 인덱스 기준
df_right_merge_index = pd.merge(df_left, df_right, left_index=True, right_index=True, how="right")

# 데이터프레임 df_left & df_right, df_right_merge_index 확인
display([df_left, df_right, df_right_merge_index], ['df_left', 'df_right', 'df_right_merge_index'])
```

```python
from IPython.display import display
import pandas as pd

# df_left, df_right 두 개의 데이터프레임 생성
df_left = pd.DataFrame({'색_left': ['red', 'orange', 'yellow', 'yellow', 'green', 'blue'],
                        '가격': [1000, 2000, 3000, 3000, 4000, 5000]},
                       index=[0, 1, 2, 3, 4, 5])

df_right = pd.DataFrame({'색_right': ['yellow', 'yellow', 'green', 'blue', 'navy', 'purple'],
                         '이름': ['노란색', '노란색', '초록색', '파란색', '남색', '보라색']},
                        index=[3, 4, 5, 6, 7, 8])

# join을 이용해 두 데이터프레임 합치기 (기본값: 인덱스)
df_right_join_index = df_left.join(df_right, how='right')

# merge를 이용해 두 데이터프레임 합치기 (인덱스 기준 설정)
df_right_merge_index = pd.merge(df_left, df_right, left_index=True, right_index=True, how="right")

# 데이터프레임 및 결괏값 확인
display([df_left, df_right], ['df_left', 'df_right'])
display([df_right_join_index, df_right_merge_index], ['df_right_join_index', 'df_right_merge_index'])
```

```python
df_1 = pd.DataFrame({'색1': ['red', 'orange'],
                     '이름1': ['빨간색', '주황색']},
                     index=[0, 1])

df_2 = pd.DataFrame({'색2': ['yellow', 'green'],
                     '이름2': ['노란색', '초록색']},
                    index=[2, 3])

df_3 = pd.DataFrame({'색3': ['blue', 'navy'],
                     '이름3': ['파란색', '남색']},
                    index=[8, 9])

# 기본값: 행(row) 기준으로 연결
result_row = pd.concat([df_1, df_2, df_3])

# 연결할 데이터프레임 확인
display([df_1, df_2, df_3], ['df_1', 'df_2', 'df_3'])

# 결과 데이터프레임
result_row
```

```python
# 열 이름을 {기존:새이름}으로 설정 뒤, 데이터프레임에 다시 저장
df_1.rename(columns={"색1": "색", "이름1":"이름"}, inplace=True)
df_2.rename(columns={"색2": "색", "이름2":"이름"}, inplace=True)
df_3.rename(columns={"색3": "색", "이름3":"이름"}, inplace=True)

# 기본값: 행 기준으로 결합
result_row = pd.concat([df_1, df_2, df_3])

# 인덱스 값 확인 필요
display([df_1, df_2, df_3, result_row], ['df_1', 'df_2', 'df_3', 'result_row'])
```

```python
# ignore_index = True 설정
result_row = pd.concat([df_1, df_2, df_3], ignore_index=True)

# 인덱스 값 확인 필요
display(df_1, df_2, df_3, result_row)
```

```python
# 가로로 데이터를 연결하는 방법
df_1 = pd.DataFrame({'색': ['red', 'orange'],
                     '이름': ['빨간색', '주황색']},
                     index=[0, 1])

df_2 = pd.DataFrame({'색': ['yellow', 'green'],
                     '이름': ['노란색', '초록색']},
                    index=[2, 3])

df_3 = pd.DataFrame({'색': ['blue', 'navy'],
                     '이름': ['파란색', '남색']},
                    index=[8, 9])

# axis = 1 열 방향을 합치기
result_col = pd.concat([df_1, df_2, df_3], axis = 1)

# 연결하려는 데이터
display(df_1, df_2, df_3)

# 결과 데이터프레임
result_col
```

```python
# 기존 인덱스 값을 지우고(drop=True), 0부터 다시 시작
df_2.reset_index(drop=True, inplace=True)
df_3.reset_index(drop=True, inplace=True)

result_col = pd.concat([df_1, df_2, df_3], axis = 1, ignore_index=False)

# 연결하려는 데이터
display(df_1, df_2, df_3)

# 결과 데이터 프레임
result_col
```

```python
# append: axis=0 (세로)
df_1 = pd.DataFrame({'색': ['red', 'oragne'],
                     '이름': ['빨간색', '주황색']},
                    index=[0, 1])

df_2 = pd.DataFrame({'색': ['yellow', 'green'],
                     '이름': ['노란색', '초록색']},
                    index=[2, 3])

df_3 = pd.DataFrame({'색': ['blue', 'navy'],
                     '이름': ['파란색', '남색']},
                    index=[8, 9])

# index 값을 초기화하여 데이터 연결
result_append = df_1.append([df_2, df_3], ignore_index=True)

display(df_1, df_2, df_3, result_append)
```

```python
# add_row라는 Series 생성
add_row = pd.Series(['purple', '보라색'], index=['색', '이름'])
print('>>> add_row 배열\n', add_row)
print('>>> 색 이름 출력:', add_row["색"])

# result_append에 .concat()을 이용해 add_row 추가
added_concat = pd.concat([result_append, add_row], ignore_index=True)

# result_append에 .append()를 이용해 add_row 추가
added_append = result_append.append(add_row, ignore_index=True)

# result_append, concat 사용, append 사용했을 때 결괏값 출력
display(result_append, added_concat, added_append)
```
